```python
'''随机重复值数字列表，去重后排序操作'''
import random

maxn = 10
a = []
for i in range(maxn):
    a.append(random.randrange(1, 7))
print("原始数据：")
print(a)

# 去除重复的项目
key, n = 0, maxn
while key < n:
    i = n - 1
    while a[i] != a[key]:
        i -= 1
    if i == key:
        key += 1
    else:
        a.remove(a[i])
        n -= 1

# 排序算法
for i in range(n):
    for j in range(len(a)-i-1):
        if a[j] > a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]
print("去重后排序数据：")
print(a)
print("共删除数据：", maxn - n, "个")
    
```

```python
'''
Author: Libre Gu 
Date: 2023-02-21 10:56:00
LastEditors: Libre Gu 
LastEditTime: 2023-05-20 08:01:42
FilePath: /PythonJupyter/Demo.ipynb
Description: 
Copyright (c) 2023 by Libre, All Rights Reserved. 
'''
string = "05/21/2021"   # 原始数据
split_char = "/"        # 分隔符
tmp = ""                # 临时拆分出的字符串
out_list = []           # 分割后数据的列表

# 遍历每一个字符
for s in string:
    # 如果当前字符不是分隔符，把字符串累加起来
    if s != split_char:
        tmp += s
    else: # 如果当前字符是分隔符，把之前累加起来的字符串添加到列表中
        out_list.append(tmp)
        #清空临时字符串变量，以便下次循环时候继续累加
        tmp = ""
# 把最后面没有处理的字符串添加进列表
if tmp != "":
    out_list.append(tmp)
print(out_list)
```

```python
# 冒泡排序
import numpy as np
L = np.random.randint(1, 100, 10)

for i in range(len(L)):
    for j in range(len(L) - i - 1):
        if L[j] > L[j+1]:
            L[j],L[j+1] = L[j+1],L[j]
print(L)
```

```python
# 选择排序
import numpy as np
L = np.random.randint(0, 100, 10)
for i in range(len(L)):
    k = i
    for j in range(i+1, len(L)):
        if L[k] > L[j]:
            k = j
        if k != i:
            L[k], L[i] = L[i], L[k]
print(L)
```
