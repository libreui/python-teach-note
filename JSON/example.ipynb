```python
import json
# 对像
cat = {
    "name": "Kitty",
    "age": 1,
    'eye-color': 'blue',
    'hair-color': 'red'
}

class Dog:
    def __init__(self):
        self.name = "大黄"
        self.age = 1
        self.eye_color = 'black'
        self.hair_color = '黄色'

dog = Dog()

json_str = json.dumps(dog.__dict__, ensure_ascii=False)
print(json_str)
print(type(json_str))

```

```python
json_str = '[3,3,3,3,3]'

dog = json.loads(json_str)
print(dog)
print(type(dog))

lst = ['a', 'b', 'c']
print(json.dumps(lst))
```
