```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(['R1', 'R2', 'R3', 'R4'])
y = np.array([38, 45, 90, 10])
plt.bar(x, y)
plt.show()
```

```python
x = np.array(['R1', 'R2', 'R3', 'R4'])
y = np.array([38, 45, 90, 10])
plt.bar(x, y, width=0.5, align='edge')
plt.show()
```

```python
x = np.array(['R1', 'R2', 'R3', 'R4'])
y = np.array([38, 45, 90, 10])
plt.barh(x, y, color="#FF0000")
plt.show()
```
