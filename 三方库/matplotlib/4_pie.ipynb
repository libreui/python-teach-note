```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([50, 48, 2])
plt.pie(y, labels=['A', 'B', 'C'], labeldistance=1.1)
plt.title('Pie Example')
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
# 连线样式
plt.pie(y,
        labels=['A','B','C','D'], # 设置饼图标签
        colors=["#d5695d", "#5d8ca8", "#65a479", "#a564c9"], # 设置饼图颜色
        shadow=True, # 阴影,
        explode=(0, 0.1, 0, 0), # 突出
        autopct="%0.1f%%", # 格式化
        startangle=230, # 旋转
       )
plt.title("Pie Example") # 设置标题
plt.show()
```
