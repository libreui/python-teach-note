```python
# 二维列表 [(x1, y1), (x2, y2), ......] N(100)
# 将点绘制在plot
import matplotlib.pyplot as plt
from random import randint
import numpy as np
count = 10
ticks = np.arange(1.0, 11.0, 0.25)
y_ticks = np.arange(0, 101, 5)
points = [(randint(1, 100), randint(1, 100)) for _ in range(count)]
# x_values = [x for x, y in points]
y_values = [y for x, y in points]

fig, ax = plt.subplots()
ax.set_xticks(ticks)
ax.set_yticks(y_ticks)

ax.plot( y_values, '-o')
ax.grid(linestyle='--', alpha=0.3)

plt.show()
```

```python
from matplotlib.axes import Axes
import matplotlib.pyplot as plt
import numpy as np
ax: list[Axes] = []
fig, ax = plt.subplots(1, 2)
y_values = np.arange(10)
ax[0].plot(y_values)
ax[0].set_title('Line')

x1_values = np.arange(10)
y1_values = np.square(x1_values)
ax[1].plot(x1_values, y1_values)
ax[1].set_title('Square func')

plt.show()
```
