```python
#  [(x1, y1), (x2, y2), ......] N(100)
# plot
import matplotlib.pyplot as plt
from random import randint
import numpy as np
count = 10
ticks = np.arange(1.0, 11.0, 0.5)
points = [(randint(1, 100), randint(1, 100)) for _ in range(count)]
# x_values = [x for x, y in points]
y_values = [y for x, y in points]
plt.plot(y_values)
plt.xticks(ticks)

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 2)
y_values = np.arange(10)
ax[0].plot(y_values)
ax[0].set_title('Line')

x1_values = np.arange(10)
y1_values = np.square(x1_values)
ax[1].plot(x1_values, y1_values)
ax[1].set_title('Square func')

plt.show()
```
