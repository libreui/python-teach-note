```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 基本散点图
x = np.random.randint(1, 100, 10)
y = np.random.randint(1, 100, 10)
print(x,y)
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.randint(1, 100, 8)
y = np.random.randint(1, 100, 8)
sizes = np.array([20,50,100,200,500,1000,60,90])
colors = np.array([10, 20, 30, 40, 50, 60, 70, 80])
plt.scatter(x, y, s=sizes, c=colors, cmap='gray_r')
plt.colorbar()
plt.show()
```
