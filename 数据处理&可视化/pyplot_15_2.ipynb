```python
# 使用scatter绘制散点图并设置样式
import matplotlib.pyplot as plt

x_value = [1, 2, 3, 4, 5]
y_value = [1, 4, 9, 16, 25]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_value, y_value, s=200)

# 设置字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签

# 设置图标标题并给坐标轴加上标签
ax.set_title("平方数", fontsize=24)
ax.set_xlabel("值", fontsize=14)
ax.set_ylabel("值的平方", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
# 15.2.6 自动计算数据
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]

plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签

fig, ax = plt.subplots()
ax.scatter(x_value, y_value, s=10)

ax.set_title("平方数", fontsize=24)
ax.set_xlabel("值", fontsize=14)
ax.set_ylabel("值的平方", fontsize=14)
ax.tick_params(axis='both', labelsize=10)

# 设置每个坐标轴的取值范围
ax.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
# 15.2.6 自动计算数据
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]

plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签

fig, ax = plt.subplots()
# 给数据点增加颜色
ax.scatter(x_value, y_value, s=10, c='red')

ax.set_title("平方数", fontsize=24)
ax.set_xlabel("值", fontsize=14)
ax.set_ylabel("值的平方", fontsize=14)
ax.tick_params(axis='both', labelsize=10)

# 设置每个坐标轴的取值范围
ax.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
# 15.2.8 使用颜色映射值
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]

plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签

fig, ax = plt.subplots()
# 使用颜色来映射值
ax.scatter(x_value, y_value, s=10, c=y_value, cmap=plt.cm.Blues)

ax.set_title("平方数", fontsize=24)
ax.set_xlabel("值", fontsize=14)
ax.set_ylabel("值的平方", fontsize=14)
ax.tick_params(axis='both', labelsize=10)

# 设置每个坐标轴的取值范围
ax.axis([0, 1100, 0, 1100000])

plt.show()
```
