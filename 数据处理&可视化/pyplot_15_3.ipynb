```python
# 15.3.3 绘制随机漫步图
import matplotlib.pylab as plt
from random_walk import RandomWalk

# 创建一个RandomWalk实例
rw = RandomWalk()
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s=15)

plt.show()
```

```python
# 15.3.4 模拟多次随机漫步
import matplotlib.pylab as plt
from random_walk import RandomWalk

while True:
    # 创建一个RandomWalk实例
    rw = RandomWalk()
    rw.fill_walk()
    plt.style.use('classic')
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s=15)

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

```python
# 15.3.5 设置随机漫步图的样式
import matplotlib.pylab as plt
from random_walk import RandomWalk

while True:
    # 创建一个RandomWalk实例
    rw = RandomWalk()
    rw.fill_walk()
    plt.style.use('classic')
    fig, ax = plt.subplots()
    
    # 设置随机点的颜色
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15)

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

```python
# 15.3.5.2 突出起点和终点
import matplotlib.pylab as plt
from random_walk import RandomWalk

# 创建一个RandomWalk实例
rw = RandomWalk()
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()

# 突出起点和终点
ax.scatter(0, 0, c='green', edgecolors='none', s=100)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100)

plt.show()

```

```python
# 15.3.5.3 隐藏坐标轴
import matplotlib.pylab as plt
from random_walk import RandomWalk

# 创建一个RandomWalk实例
rw = RandomWalk()
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()

# 突出起点和终点
ax.scatter(0, 0, c='green', edgecolors='none', s=100)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100)

# 隐藏坐标轴
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.show()
```

```python
# 15.3.5.4 增加点数
import matplotlib.pylab as plt
from random_walk import RandomWalk

# 创建一个RandomWalk实例
rw = RandomWalk(50_000)
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()

ax.scatter(rw.x_values, rw.y_values, c=range(rw.num_points), cmap=plt.cm.Blues, edgecolors='none', s=1)

plt.show()
```

```python
# 15.3.5.5 调整尺寸以适合屏幕
import matplotlib.pylab as plt
from random_walk import RandomWalk

# 创建一个RandomWalk实例
rw = RandomWalk(50_000)
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(10, 6), dpi=128)

ax.scatter(rw.x_values, rw.y_values, c=range(rw.num_points), cmap=plt.cm.Blues, edgecolors='none', s=1)

plt.show()
```
