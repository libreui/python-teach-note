```python
print("%d * %d = %d" % (1, 1, 1))

# format 格式化输出
print("I am {0}, age {1}".format('Libre', 40))
print("{0} * {0} = {0}".format(1))

# 用变量名对应值替换
print("我得名字叫{name}, 年龄{age}, 性别{sex}, 他也叫{name}".format(age=40,name='Libre', sex='男'))


print("{}, {}, {}".format(1, 1, 1))

# 补位
print("%10s" % ('A'))
print("{:<10}B".format('A'))
```

```python
# format格式化
names_0 = ["唐三", "小舞", "比比东"]
names_1 = ["萧炎", "美杜莎", "熏儿"]

print("我叫{0[1]}, 她叫{1[1]}".format(names_0, names_1))

```

```python
# format 遇到了数字
import math

print("{:.2f}".format(math.pi))
print("{:.0f}".format(math.pi))

# 数字补位
print("{:0^9}".format(5))
print("{:*^9}".format(5))
print("{:-^9}".format(5))
print("{:A^9}".format(5))

# 字符串补位
print("{:-^50}".format("Hello"))
```

```python
# 正负数、百分号格式化
print("{:+d}".format(5))
print("{:-f}".format(5))

print("{:,}".format(1000000000))
num = 1_000_000_000
print(num)
```
