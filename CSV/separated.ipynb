```python
my = {
    "name": "张三",
    "hp": 98,
    "mp": 56,
    "atk": 10,
    "def": 10,
    "spd": 10,
    "luk": 10,
    "exp": 39878,
}

with open("my.csv", "w") as f:
    save = [str(my['exp']), str(my['hp']), str(my['mp'])]
    f.write(",".join(save))

he = {
    "exp": 0,
    "hp": 0,
    "mp": 0,
}

with open("my.csv", "r") as f:
    p = f.read().strip().split(',')
    he['exp'] = int(p[0])
    he['hp'] = int(p[1])
    he['mp'] = int(p[2])
print(he)
```

```python
# -*- coding: utf-8 -*-
import csv
c = [
    ['张三', '95', '98', '78', '65'],
    ['李四', '90', '92', '88', '76'],
    ['王五', '85', '90', '80', '70'],
]
with open('cj.csv', 'w', newline="") as f:
    w = csv.writer(f)
    w.writerows(c)
print(c)
```

```python
# 如何将这个csv文件中的数据还原
c = []
with open("cj.csv", "r") as f:
    for row in f:
        c.append(row.strip().split(','))
print(c)
```
