````python
# 通用文件操作
# 边出数据边写入文件
with open("jj.csv", "w", newline="", encoding="utf-8") as f:
    for i in range(1, 10):
        row = []
        for j in range(1, i+1):
            row.append("{0}x{1}={2}".format(j, i, i*j))
        f.write(" ".join(row))
        f.write("\n")
print('OK')
````

3. 生成100个随机密码，长度3-12、随机大小写字母数字组合，写入file文件夹下名字叫password.csv的文件中，一行一个密码

````python
import random
passwords = []
for i in range(100):
    password = ''
    for j in range(3, random.randint(3, 12)):
        index = random.randint(0, 2)
        if index == 0:
            password += chr(random.randint(65, 90))
        elif index == 1:
            password += chr(random.randint(97, 122))
        else:
            password += str(random.randint(0, 9))
    passwords.append(password)
print(passwords)
````

````python
import random
A = [chr(x) for x in range(65, 91)]
A += [chr(x) for x in range(97, 123)]
A += [str(x) for x in range(10)]
passwords = []
for n in range(100):
    password = ''
    for i in range(3, random.randint(3, 13)):
        password += random.choice(A)
    passwords.append(password)
print(passwords)
````

````python
import csv

data = []
with open('file/user.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    data = list(reader)

chinese_data = {
    'best_student': '',
    'max': 0,
    'min': 99999,
    'low': 0,
    'count': 0,
    'sum': 0
}

# 数据处理
for item in data:
    #统计语文总分
    chinese_data['sum'] += int(item['yuwen'])
    # 及格人数
    if int(item['yuwen']) > 60:
        chinese_data['low'] += 1
    # 最大分数
    if int(item['yuwen']) > chinese_data['max']:
        chinese_data['max'] = int(item['yuwen'])
        chinese_data['best_student'] = item['name']
    # 最大分数
    if int(item['yuwen']) < chinese_data['min']:
        chinese_data['min'] = int(item['yuwen'])

chinese_data['count'] = len(data)
                            
print(chinese_data)
````
