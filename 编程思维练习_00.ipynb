### 编程思维练习
- 计算并输出1到100之间的所有偶数的和。
- 实现字符串的反转功能，例如输入"Hello, World!"，输出"!dlroW ,olleH"。
- 从一个整数列表中找出最大值和最小值，并计算它们的差值。
- 判断一个给定的字符串是否为回文字符串，即正着读和反着读都一样，例如输入"level"，输出True。
- 模拟一个简单的猜数字游戏，随机生成一个目标数字，用户通过输入猜测的数字来猜测目标数字，直到猜中为止。
- 实现一个简单的计算器，支持加法、减法、乘法和除法运算，用户输入运算符和操作数后，输出计算结果。
- 实现一个简单的待办事项清单应用，支持添加、删除和展示待办事项，并可以标记已完成的事项。
- 实现一个简单的短信加密解密工具，可以将输入的短信内容进行加密，并可以输入密钥进行解密。
- 读取一个文本文件，统计其中每个单词出现的次数，并按照出现次数从高到低的顺序输出。
- 实现一个简单的计算器，可以进行基本的数学运算（如加法、减法、乘法、除法），并支持括号的使用。
